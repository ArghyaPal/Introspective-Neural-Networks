```python
import torch
from sampling import Synthesis
```

```python
class toy_example(torch.nn.Module):
    def __init__(self, size=(10, 1, 5, 5), std=0.3):
        super(toy_example, self).__init__()
        # parameter for sampling
        self.X = torch.nn.Parameter(torch.empty(size).normal_(mean=0, std=std));
        self.X.requires_grad = False;
        
        # other net parameters
        self.Y = torch.nn.Parameter(torch.empty(size).normal_(mean=0, std=std));
    
    def forward(self):
        return 0;
    
    def to_synth(self):
        '''
        Multi-dimensional parabola function with leaves going down. 
        During maximization should converge to zero
        '''
        return -((self.X*self.Y)**2).sum(dim=(-1, -2));
```

```python
m = toy_example();
s = Synthesis(init_std=0.3);
s.sample(m, num_iter=5000, learning_rate=1e-1)
```
